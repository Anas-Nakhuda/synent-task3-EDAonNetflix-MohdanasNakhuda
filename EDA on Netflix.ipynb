```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df=pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.tail()
```

```python
print("Shape:", df.shape)
display(df.head())
print("\nColumns:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum().sort_values(ascending=False))
```

```python
df=df.drop_duplicates()
```

```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df.columns
df.head()
```

```python
for col in ["director", "cast", "country", "rating", "duration", "listed_in"]:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

```python
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month_name()
```

```python
df["duration_min"] = np.where(
    df["type"] == "Movie",
    pd.to_numeric(df["duration"].str.extract(r"(\d+)")[0], errors="coerce"),
    np.nan
)
print("After cleaning, shape:", df.shape)
print("\nMissing values after cleaning:\n", df.isnull().sum().sort_values(ascending=False))
```

```python
print("\nType counts:\n", df["type"].value_counts())
print("\nTop ratings:\n", df["rating"].value_counts().head(10))
print("\nRelease year range:", df["release_year"].min(), "-", df["release_year"].max())
print("\nMovie duration stats:\n", df["duration_min"].describe())
```

```python
# Visual 1: Movies vs TV Shows
plt.figure()
sns.countplot(data=df, x="type", palette="Set2")
plt.title("Distribution of Content Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
# Visual 2: Top 10 countries
country_series = (
    df["country"]
    .str.split(", ")
    .explode()
    .value_counts()
    .head(10)
)
plt.figure()
sns.barplot(x=country_series.values, y=country_series.index, palette="Blues_r")
plt.title("Top 10 Countries by Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
# Visual 3: Content added by year
year_trend = df["year_added"].value_counts().sort_index()
plt.figure()
sns.lineplot(x=year_trend.index, y=year_trend.values, marker="o")
plt.title("Netflix Content Added Over Years")
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.show()
```

```python
# Visual 4: Top genres
genre_series = (
    df["listed_in"]
    .str.split(", ")
    .explode()
    .value_counts()
    .head(10)
)
plt.figure()
sns.barplot(x=genre_series.values, y=genre_series.index, palette="viridis")
plt.title("Top 10 Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
# Visual 5: Rating distribution
top_ratings = df["rating"].value_counts().head(10)
plt.figure()
sns.barplot(x=top_ratings.index, y=top_ratings.values, palette="magma")
plt.title("Top Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visual 6: Movie duration histogram
plt.figure()
sns.histplot(df["duration_min"].dropna(), bins=30, kde=True, color="teal")
plt.title("Distribution of Movie Duration (minutes)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```
